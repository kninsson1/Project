```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
dt= xr.open_dataset('Africa_cru_data.nc')
dt
```

```python
pre_dt=dt['pre']
pre_dt
```

```python
llt = pre_dt.sel(lon=slice(-3.5, 1.5), lat=slice(4.5, 11.5), time=slice('1983', '1993'))
llt
```

```python
Annual_Where = llt.where(llt>250).groupby('time.year')
```

```python
Annual_Count=Annual_Where.count('time')
Annual_Count
```
